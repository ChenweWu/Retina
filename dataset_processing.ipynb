```python
import pandas as pd
import os
import shutil
import glob
import splitfolders
```

```python
fpath_csv = 'labels.csv'
df_csv = pd.read_csv(fpath_csv)
df_csv.head()
```

```python
# 1 for male or female?1 male 2 female okok

df_image_male = df_csv[df_csv['patient_sex']==1]['image_id']
df_image_female = df_csv[df_csv['patient_sex']==2]['image_id']

df_image_male.head()
```

```python
path_dest = './pre_data'
if not os.path.exists(path_dest):
    os.makedirs(path_dest)

for gender in ['male', 'female']:
    path_gender = os.path.join(path_dest, gender)
    if not os.path.exists(path_gender):
        os.makedirs(path_gender)
```

```python
path_image = './images'

fpath_images = glob.glob(path_image + '/*.jpg')

for fpath in fpath_images:
    fname = os.path.basename(fpath) # img0002.jpg
    fname_ = os.path.splitext(fname)[0] # img00002
    
    if fname_ in df_image_male.values:
        # copy the file
        fpath_dst = os.path.join(path_dest, 'male', fname)
        shutil.copyfile(fpath, fpath_dst)

    elif fname_ in df_image_female.values:
        fpath_dst = os.path.join(path_dest, 'female', fname)
        shutil.copyfile(fpath, fpath_dst)

    else:
        assert False
```

```python
# input = "image"
# output = "output" #where you want the split datasets saved. one will be created if it does not exist or none is set

splitfolders.ratio('pre_data', output="data", seed=1337, ratio=(.9, .0, .1)) # ratio of split are in order of train/val/test. You can change to whatever you want. For train/val sets only, you could do .75, .25 for example.
```
